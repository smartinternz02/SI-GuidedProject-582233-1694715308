##Univariate analysis

```python
plt.hist(df['lot_area_renov'], bins=20)
plt.title('Histogram of lot_area_renov')
plt.show()
```

```python
category_counts = df['Number of schools nearby'].value_counts()

plt.figure(figsize=(8, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Bar Plot of Number of schools nearby')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Distance from the airport'], vert=True)
plt.title('Box Plot of Distance from the airport')
plt.show()
```

##Bi varaiate analysis

```python
plt.scatter(df['Price'], df['Distance from the airport'])
plt.xlabel('Price')
plt.ylabel('Distance from the airport')
plt.title('Scatter Plot between Price and Distance from the airport')
plt.show()
```
